```python
import json
```

```python
emoji_json = json.load(open("emoji_json.json", "r"))
```

```python
flipped = {}
emoji_count = {} 
for emoji, words_string in emoji_json.items(): 
    words = words_string.split(" | ")
    emoji_count[emoji] = len(words)
    for word in words: 
        word = word.lower()
        to_append = (emoji, emoji_count[emoji])
        if word in flipped: 
            flipped[word].append(to_append)
        else: 
            flipped[word] = [to_append]
            
for word in flipped.keys(): # count how many words trigger each emoji and prioritize the ones with fewer
    flipped[word] = sorted(flipped[word], key=lambda pair: pair[1], reverse=False)
    flipped[word] = [pair[0] for pair in flipped[word]]
```

```python
# make secondary copy of original json with modifications 
# ik this is a dumb way to do it but lazy 
emoji_to_text = {}
for word in flipped.keys(): 
    for emoji in flipped[word]: 
        to_append = [word, len(flipped[word])]
        if emoji in emoji_to_text: 
            emoji_to_text[emoji].append(to_append)
        else: 
            emoji_to_text[emoji] = [to_append]
for emoji in emoji_to_text.keys(): # prioritize terms that have longest list of emojis associated w them
    emoji_to_text[emoji] = sorted(emoji_to_text[emoji], key=lambda pair: pair[1], reverse=True)
    emoji_to_text[emoji] = [pair[0] for pair in emoji_to_text[emoji]]
```
